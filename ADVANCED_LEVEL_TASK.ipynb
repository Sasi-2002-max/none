```python
import plotly.express as px # Import the plotly.express module and assign it to the variable 'px'
import pandas as pd

df = pd.read_csv('/content/netflix_titles.csv', on_bad_lines='skip')   # Read the CSV file 'your_data.csv' into a pandas DataFrame and assign it to the variable 'df'

x=df.groupby(['rating']).size().reset_index(name='counts')
print(x)

piechart = px.pie(x, values='counts',names='rating',title='distribution rationg of content')
piechart.update_layout(width=700, height=600) # Update the layout of the pie chart to set the figure size
piechart.show()
```

```python
df['director']=df['director'].fillna('No Director Specified')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
directors_list = pd.DataFrame()
print(directors_list)
```

```python
directors_list = df['director'].str.split(',', expand=True).stack()
print(directors_list)
```

```python
directors_list= directors_list.to_frame()
print(directors_list)
```

```python
directors_list.columns = ['directors']
print(directors_list)
```

```python
directors = directors_list.groupby(['directors']).size().reset_index(name='Total Count')
print(directors)
```

```python
directors = directors[directors.directors != 'No Director Specified']
print(directors)
```

```python
directors = directors.sort_values(by=['Total Count'], ascending=False)
print(directors)
```

```python
top5directors = directors.head()
```

```python
print(top5directors)
```

```python
top5directors=top5directors.sort_values(by=['Total Count'])
barchart = px.bar(top5directors, x='Total Count', y='directors', title='Top 5 Directors With The Most Content')
barchart.update_layout(width=700, height=600)
barchart.show()
```

```python
df['cast'] = df['cast'].fillna('No Cast Specified')
cast_df = pd.DataFrame()
cast_df = df['cast'].str.split(',', expand=True).stack()
cast_df = cast_df.to_frame()
cast_df.columns = ['Actor']
actors = cast_df.groupby(['Actor']).size().reset_index(name='Total Count')
actors = actors.sort_values(by=['Total Count'], ascending=False)
top5actors = actors.head()
top5actors = top5actors.sort_values(by=['Total Count'])
barchart2 = px.bar(top5actors, x='Actor', y='Total Count', title='Top 5 Actors With The Most Content')
barchart2.update_layout(width=700, height=600)
barchart2.show()
print(cast_df)

```

```python
df1 = df[['type','release_year']]
df1= df1.rename(columns = {"release_year": "Release Year"})
df1.head()
```

```python
df2=df1.groupby(['Release Year','type']).size().reset_index(name='Total Count')
print(df2)
```

```python
df2.head()
```

```python
df2=df2[df2["Release Year"]>=2000]
graph = px.line(df2, x='Release Year', y='Total Count', color='type', title='Trend of content produced over the years')
graph.update_layout(width=700, height=600)
graph.show()
```

```python
df3 = df[['release_year','description']]
df3 = df3.rename(columns = {'release_year':'Release Year','description':'Description'})
for index, row in df3.iterrows():
  d=row['Description']
  testimonial = TextBlob(d)
  p=testimonial.sentiment.polarity
  if p==0:
    sent = 'Neutral'
  elif p>0:
    sent = 'Positive'
  else:
    sent = 'Negative'
  df3.loc[[index, 2],'Sentiment']=sent

df3 = df3.groupby(['Release Year','Sentiment']).size().reset_index(name='Total Count')

df3=df3[df3['Release Year']>=2005]
bargraph = px.bar(df3, x='Release Year', y='Total Count', color='Sentiment', title='Sentiment of content produced over the years')
bargraph.update_layout(width=700, height=600)
bargraph.show()
```
